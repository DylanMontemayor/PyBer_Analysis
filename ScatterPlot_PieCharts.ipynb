```python
import matplotlib.pyplot as plt
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]#independent
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]#dependent
```

```python
plt.scatter(x,y, c="green", marker="x",s=x)
plt.xlabel("Total number of rides (Per city)")
plt.ylabel("Average Fare($)")
plt.title("Average Fare vs Total Rides")
plt.show
```

```python
categories = ["Pants", "Shirts", "Outwear", "Pajamas","Underwear", "Shoes"]
```

```python
revenue = [25,15,5,20,10,25]
```

```python
explode=[0,0.1,0,0,0,0]
```

```python
plt.pie(revenue, labels=categories, autopct="%1.1f%%", explode=explode)
plt.show()
```
