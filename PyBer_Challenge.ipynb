# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_percity=pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_percity
```

```python
# 2. Get the total drivers for each city type
drivers_percity=city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_percity
```

```python
#  3. Get the total amount of fares for each city type
fares_percity=pyber_data_df.groupby(["type"]).sum()["fare"]
fares_percity
```

```python
#  4. Get the average fare per ride for each city type. 
fare_average_percity=fares_percity/rides_percity
fare_average_percity
```

```python
# 5. Get the average fare per driver for each city type. 
fare_average_perdriver=fares_percity/drivers_percity
fare_average_perdriver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary=pd.DataFrame({
    "Total Rides":rides_percity,
    "Total Drivers":drivers_percity,
    "Total Fares":fares_percity,
    "Average Fare per Ride":fare_average_percity,
    "Average Fare per Driver":fare_average_perdriver,
})
pyber_summary
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary.index.name = None
```

```python
#  8. Format the columns.

pyber_summary["Total Rides"] = pyber_summary["Total Rides"].map("{:,}".format)
pyber_summary["Total Drivers"] = pyber_summary["Total Drivers"].map("{:,}".format)
pyber_summary["Total Fares"] = pyber_summary["Total Fares"].map("${:,.2f}".format)
pyber_summary["Average Fare per Ride"] = pyber_summary["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary["Average Fare per Driver"] = pyber_summary["Average Fare per Driver"].map("${:.2f}".format)
pyber_summary
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
newpyber_data_df=pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
newpyber_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
newpyber_data_df = newpyber_data_df.reset_index()
newpyber_data_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
newpyber_data_pivot=newpyber_data_df.pivot(index="date",columns="type",values="fare")
newpyber_data_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_date_df=newpyber_data_pivot.loc[(newpyber_data_pivot.index>='2019-01-01')&
(newpyber_data_pivot.index<='2019-04-29')]
pyber_date_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_date_df.index = pd.to_datetime(pyber_date_df.index)
pyber_date_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_date_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_date_week_df=pyber_date_df.resample("W").sum()
pyber_date_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.dates as mdates
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot()
    
ax.plot(pyber_date_week_df.index, pyber_date_week_df["Urban"], marker="_", color="gold", linewidth=2, label='Urban')
ax.plot(pyber_date_week_df.index, pyber_date_week_df["Suburban"], marker="_", color="coral", linewidth=2, label='Suburban')
ax.plot(pyber_date_week_df.index, pyber_date_week_df["Rural"], marker="_", color="skyblue", linewidth=2, label='Rural')

ax.xaxis.set_major_formatter(
    mdates.ConciseDateFormatter(ax.xaxis.get_major_locator()))
ax.set_xlabel("Date")
ax.set_ylabel("Fare($USD)")
ax.set_title("Total Fare by City Type")
ax.legend(mode="Expanded",title="City Types",loc=(0.5,0.5))

ax.grid
#Save the figure.
plt.savefig("analysis/FigChallenge.png")

# Show the plot
plt.show()
```
