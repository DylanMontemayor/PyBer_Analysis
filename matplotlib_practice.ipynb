```python
import matplotlib.pyplot as plt
```

```python
#Create a plot usint the MatLab Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create a Line Chart Using the Object-Oriented Interface
#Create it with ax.plt()
fig, ax=plt.subplots()
#fig=plt.figure()
ax.plot(x_axis, y_axis)
#ax=fig.add_subplot()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

#ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

#xlim
#plt.xlim("Jan","June")
#plt.xlim(0,5)

#figure size
#plt.figure(figuresize=(20,10))
#plt.subplot(figsize=(20,10))
```

```python
#Using Object-Oriented Approach
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.set_ylim(0,45)
ax.grid
```

```python
#Pie Charts MATLAB
```

```python
#Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#explote the number 7 month
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
```

```python
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.show()
```

```python
#Pie Chart Using the Object-Oriented Interface
```

```python
#Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#explote the number 7 month
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis,
       explode=explode_values, 
       colors=colors,autopct='%.1f%%',
       shadow=True,
       startangle=95,
       counterclock=False)
#ax.invert_xaxis()
plt.show()
```
